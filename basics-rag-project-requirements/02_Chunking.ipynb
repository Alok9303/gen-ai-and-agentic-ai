```python
text = '''Hometown Cha-Cha-Cha is a 2021 South Korean romantic comedy drama television series starring Shin Min-a, Kim Seon-ho and Lee Sang-yi. It is a remake of 2004 South Korean film Mr. Handy, Mr. Hong.[6] It aired on tvN from August 28 to October 17, 2021, every Saturday and Sunday at 21:00 (KST).[7][8] It is also available for streaming on Netflix.[9]

The series was a commercial hit and became one of the highest-rated dramas in Korean cable television history.[10][11] It ranked first place during its entire run for eight weeks, and the last episode achieved 12.665% nationwide rating, with over 3.2 million views.[12] It also became one of Netflix's most-watched non-English television shows, and one of its longest-running hits as it spent 16 weeks in global top ten rankings.
'''
```

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(separator = '\n',chunk_size = 200, chunk_overlap = 30)  #we can give only one separator.
```

```python
dir(splitter)
```

```python
chunks = splitter.split_text(text)
chunks
```

```python
len(chunks)
```

```python
len(chunks[0]), len(chunks[1])  #it exceeded my chunk_size. because separator='\n'. And its not recursive.
```

```python
#So to avoid this chunk_size exceeding problem, we use 'RecursiveCharecterTextSplitter'
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
r_splitter = RecursiveCharacterTextSplitter(separators=['\n\n','\n',' '],  #multiple separators. which are recursive.
                                            chunk_size = 200, # size of the chunk to be created.
                                            chunk_overlap = 30,  #size of overlapping between chunks, to maintain the context.
                                            length_function = len  #function to calculate the size of the chunk. currently we are using len of string function.
                                           )
```

```python
dir(r_splitter)
```

```python
chunks = r_splitter.split_text(text)
chunks
```

```python
len(chunks)
```

```python
print(chunks[0])
```

```python
for chunk in chunks:
    print(len(chunk))
```

So now all the chunks are below my chunk_size(200).
