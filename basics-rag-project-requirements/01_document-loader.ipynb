CSV file loader

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
```

```python
loader = CSVLoader(file_path = "healthcare_patient_records.csv")
```

```python
docs = loader.load()
```

```python
type(docs)   #'docs' is saved as a 'list'. elements of this 'docs' list are in langchain documents format.
```

```python
docs[0]
```

```python
type(docs[0])
```

```python
docs[0].metadata
```

```python
docs[0].metadata['row']   #matadata is in dictionary format.
```

```python
docs[0].metadata['source']
```

```python
docs[0].page_content  #page_content is in string format.
```

```python
docs[0].page_content[:10]
```

URL file Loader

```python
from langchain_community.document_loaders import UnstructuredURLLoader
```

```python
loader = UnstructuredURLLoader(urls = ["https://codebasics.io/courses/bootcamp/3/gen-ai-agentic-ai/lecture/3264"])
```

```python
docs = loader.load()
```

If we run this on windows + jupyter it crashes again and again becuase it is too heavy.

So dont use UnstructuredURLLoader . Instead use WebBaseLoader

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
loader = WebBaseLoader(web_path = ["https://www.cnbc.com/quotes/AAPL"])
```

```python
docs = loader.load()
```

```python
type(docs)
```

```python
docs[0].metadata   #its a dictionary
```

```python
docs[0].metadata['source']
```

```python
docs[0].page_content   #its a string.
```

```python
docs[0].page_content[:4]
```
